```python
import gym
import numpy as np
```

```python
env= gym.make('CartPole-v1')
```

```python
env.observation_space
```

```python
states_low= env.observation_space.low
states_high= env.observation_space.high
```

```python
step_0= (states_high[0]-states_low[0])/5
step_1= (states_high[1])/5
step_2= (states_high[2]-states_low[2])/5
step_3= (states_high[3])/5
```

```python
states_0=[]
states_1=[]
states_2=[]
states_3=[]
# for i in range(1000):
# #     states_0.append((states_low[0])+(i*step_0))
#     states_1.append((states_low[1])+(i*step_1))
# #     states_2.append((states_low[2])+(i*step_2))
#     states_3.append((states_low[3])+(i*step_3))
    
# for i in range(10):
#     states_0.append((states_low[0])+(i*step_0))
# #     states_1.append((states_low[1])+(i*step_1))
# #     states_2.append((states_low[2])+(i*step_2))
# #     states_3.append((states_low[3])+(i*step_3))


for i in range(5*5*5*5):
    states_0.append((states_low[0])+(i*step_0))
    states_1.append((states_low[1])+(i*step_1))
    states_2.append((states_low[2])+(i*step_2))
    states_3.append((states_low[3])+(i*step_3))
    
```

```python
def digitize(state):
    
    state_0_dig= np.digitize(state[0], bins=states_0)
    state_1_dig= np.digitize(state[1], bins=states_1)
    state_2_dig= np.digitize(state[2], bins=states_2)
    state_3_dig= np.digitize(state[3], bins=states_3)
    
    return (state_0_dig, state_1_dig, state_2_dig, state_3_dig)
    
    
```

```python
print(digitize(env.reset()))
```

```python
#Now we have digitized the state space
# 100000x100000x100000x100000

Q= np.zeros((5*5*5*5,2))
e= 0.1
alpha= 0.2
gamma=0.99
EP= 100000

for ep in range(EP):
    Tr=0
    done= False
    
    state= env.reset()
    state= digitize(state)
    
    statenum= state[0]*state[1]*state[2]*state[3]
    
    P= [e/2,e/2]
    greedy_arm= np.argmax(Q[statenum])
    
    P[greedy_arm]+=(1-e)
    action= np.random.choice([0,1], p=P)
    
    while not done:
        
        state_, reward, done, info= env.step(action)
        state_= digitize(state_)
        state_num= state_[0]*state_[1]*state_[2]*state_[3]

        Tr+=reward

        
        best_action= np.argmax(Q[state_num])
        
        
        Q[statenum][action]= Q[statenum][action]+ alpha*(reward + (gamma*Q[state_num][best_action]) - Q[statenum][action])
        
        P=[e/2,e/2]
        greedy_action= np.argmax(Q[state_num])
        P[greedy_action]+=(1-e)
        action= np.random.choice([0,1], p=P)
        
        state= state_
    
    print(ep,':',Tr, end='\r')
    
```

```python
#Q-Table of size theta_state_size*theta_dot_state_size*env.action_space.n
import math
theta_minmax = env.observation_space.high[2] 
theta_dot_minmax = math.radians(50)
theta_state_size = 50
theta_dot_state_size = 50
Q= np.random.randn(theta_state_size,theta_dot_state_size,env.action_space.n)

# For stats
ep_rewards = []
ep_rewards_table = {'ep': [], 'avg': [], 'min': [], 'max': []}

def discretised_state(state):
	#state[2] -> theta
	#state[3] -> theta_dot
	discrete_state = np.array([0,0])		#Initialised discrete array
	theta_window =  ( theta_minmax - (-theta_minmax) ) / theta_state_size
	discrete_state[0] = ( state[2] - (-theta_minmax) ) // theta_window
	discrete_state[0] = min(theta_state_size-1, max(0,discrete_state[0]))

	theta_dot_window =  ( theta_dot_minmax - (-theta_dot_minmax) )/ theta_dot_state_size
	discrete_state[1] = ( state[3] - (-theta_dot_minmax) ) // theta_dot_window
	discrete_state[1] = min(theta_dot_state_size-1, max(0,discrete_state[1]))

	return tuple(discrete_state.astype(np.int))
```

```python
#Now we have digitized the state space
# 100000x100000x100000x100000

e= 0.1
alpha= 0.2
gamma=0.99
EP= 10000

for ep in range(EP):
    Tr=0
    done= False
    
    state= env.reset()
    state= discretised_state(state)
    
    
    P= [e/2,e/2]
    greedy_arm= np.argmax(Q[state[0]][state[1]])
    
    P[greedy_arm]+=(1-e)
    action= np.random.choice([0,1], p=P)
    
    while not done:
        
        state_, reward, done, info= env.step(action)
        state_= discretised_state(state_)

        Tr+=reward

        
        best_action= np.argmax(Q[state_[0]][state_[1]])
        
        
        Q[state[0]][state[1]][action]= Q[state[0]][state[1]][action]+ alpha*(reward + (gamma*Q[state_[0]][state_[1]][best_action]) - Q[state[0]][state[1]][action])
        
        P=[e/2,e/2]
        greedy_action= np.argmax(Q[state_[0]][state_[1]])
        P[greedy_action]+=(1-e)
        action= np.random.choice([0,1], p=P)
        
        state= state_
    
    print(ep,':',Tr, end='\r')
    
```
